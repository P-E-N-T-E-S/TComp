```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files

uploaded = files.upload()

df = pd.read_csv(list(uploaded.keys())[0])

def plot_quadratico(df):
    plt.figure(figsize=(12, 6))
    plt.scatter(df['Tamanho da lista'], df['Tempo Médio (s)'], color='skyblue', alpha=0.7, edgecolor='black', label='Pontos Individuais')

    z = np.polyfit(df['Tamanho da lista'], df['Tempo Médio (s)'], 2)
    p = np.poly1d(z)

    x_fit = np.linspace(df['Tamanho da lista'].min(), df['Tamanho da lista'].max(), 100)
    plt.plot(x_fit, p(x_fit), color='red', linestyle='--', linewidth=2, label='Linha de Tendência Quadrática')

    plt.title('Relação Quadrática entre Tamanho da Lista e Tempo de Execução', fontsize=16, weight='bold')
    plt.xlabel('Tamanho da Lista', fontsize=14)
    plt.ylabel('Tempo Médio (s)', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_quadratico(df)
```
