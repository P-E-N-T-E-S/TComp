```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files

uploaded = files.upload()

df = pd.read_csv(list(uploaded.keys())[0])

def plot_quadratico(df):
    plt.figure(figsize=(12, 6))
    plt.scatter(df['Tamanho da lista'], df['Tempo Médio (s)'], color='skyblue', alpha=0.7, edgecolor='black', label='Pontos Individuais')

    z = np.polyfit(df['Tamanho da lista'], df['Tempo Médio (s)'], 2)
    p = np.poly1d(z)

    x_fit = np.linspace(df['Tamanho da lista'].min(), df['Tamanho da lista'].max(), 100)
    plt.plot(x_fit, p(x_fit), color='red', linestyle='--', linewidth=2, label='Linha de Tendência Quadrática')

    plt.title('Relação Quadrática entre Tamanho da Lista e Tempo de Execução', fontsize=16, weight='bold')
    plt.xlabel('Tamanho da Lista', fontsize=14)
    plt.ylabel('Tempo Médio (s)', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_quadratico(df)
```

```python
def calcula_media_tempo(df):
    media_df = df.groupby('Tamanho da lista')['Tempo Médio (s)'].mean().reset_index()
    return media_df

def plot_barras_media_com_tendencia(df):
    media_df = calcula_media_tempo(df)

    fig, ax = plt.subplots(figsize=(10, 6))
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
    bars = ax.bar(media_df['Tamanho da lista'], media_df['Tempo Médio (s)'], color=colors, edgecolor='black')

    for bar, tempo in zip(bars, media_df['Tempo Médio (s)']):
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.02, f'{tempo:.2f}',
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

    z = np.polyfit(media_df['Tamanho da lista'], media_df['Tempo Médio (s)'], 2)
    p = np.poly1d(z)

    x_fit = np.linspace(media_df['Tamanho da lista'].min(), media_df['Tamanho da lista'].max(), 100)
    ax.plot(x_fit, p(x_fit), color='red', linestyle='--', linewidth=2, label='Linha de Tendência Quadrática')

    ax.set_xlabel('Tamanho da Lista', fontsize=14)
    ax.set_ylabel('Tempo Médio (s)', fontsize=14)
    ax.set_title('Média do Tempo por Tamanho da Lista com Linha de Tendência', fontsize=16, fontweight='bold', pad=20)

    ax.set_ylim(0, media_df['Tempo Médio (s)'].max() * 1.4)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_linewidth(0.8)
    ax.spines['bottom'].set_linewidth(0.8)
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    ax.legend()

    plt.tight_layout()
    plt.show()

plot_barras_media_com_tendencia(df)
```
